Plot atomic radius vs. group for each period in one diagram with scatters connected by lines. (30 points) 

Hint: use ``pivot`` function to reshape the dataframe; or use ``groupby`` operation / ``for`` loop to deal with different periods. (Other methods may also work.)

```python
data22 = data.dropna(subset=['Period','AtomicRadius','Group'])
```

```python
data_atgrp = data22.pivot(index='Group',columns='Period',values ='AtomicRadius')
```

```python
data_atgrp
```

```python
plt.plot(data_atgrp,'-o')
plt.xlabel('Group')
plt.ylabel('AtomicRadius')
plt.legend(['Period 1','Period 2','Period 3','Period 4','Period 5','Period 6'])
plt.title('Plot of AtomicRadius vs Group based on Period (Vinsensius)')
plt.show()
```

Plot first ionization vs. electronegativity for each period in different subplots, with the size and color of scatters correlated to the atomic radius. (30 points)
* Create a plot with 6 subplots
* In each subplot, create scatter plot for first ionization vs. electronegativity for each period (subplot 1 for period 1 etc.) with the size and color of scatters correlated to the atomic radius

Hint: use ``groupby`` and ``get_group`` functions to extract data with desired period. (Other methods may also work.)

```python
data33 = data.dropna(subset=['Period','AtomicRadius','FirstIonization','Electronegativity' ])
```

```python
data_period = data33.groupby('Period')
```

```python
data_period.get_group(2)[['FirstIonization','Electronegativity']]
```

```python
fig, axes = plt.subplots(3, 2, sharex='col', sharey='row', figsize=(15, 18))
plt.rc('axes', labelsize=16)    # fontsize of the x and y labels
for i, ax in enumerate(axes.flatten()):
    im = ax.scatter(data_period.get_group(i+1)['Electronegativity'],data_period.get_group(i+1)['FirstIonization'], 
                   s=data_period.get_group(i+1)['AtomicRadius']*100, c=data_period.get_group(i+1)['AtomicRadius'],cmap='viridis')
    ax.set(xlabel='Electronegativity', ylabel='FirstIonization')
    ax.set_title(('Period '+str(i+1)),size=14)
    ax.grid()
    cbar = plt.colorbar(im, ax=ax)
    cbar.set_label('Atomic Radius',size=15)

plt.suptitle('Plot of FirstIonization vs Electronegativity for Different Periods (Vinsensius)',size=17)
fig.tight_layout(pad=3.0)
plt.show()
```
